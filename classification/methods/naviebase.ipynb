```python
import numpy as np
import pandas as pd
import os 
import re
class NaiveBaye():
    # 数据的结构必须是 最后一行是作者，倒数第二行是文本

    def __init__(self, train_fileName):
        # 1 读取数据
        self.train_data = pd.read_csv(train_fileName)
        self.wordTable = self.getWordTable()  # 词表
        self.authorWordTable = self.getAuthorWordTable()

    def getWordTable(self):  # 总的词表
        wordTable = set([])
        for text in self.train_data["text"]:
            text = re.split("[,.?，。？;\"\'\“\”\‘\’!！ ]",text) #利用正则表达式对句子分词
            wordTable = wordTable | set(text)
        return list(wordTable)
    
    def getAuthorWordTable(self):
        # 表的形式如：{"":[],"":[],"":[]}
        authors = self.train_data["author"].unique()
        author_wordTable = {}
        for author in authors:
            author_wordTable[author] = [1]*len(self.wordTable)
        for text, author in zip(self.train_data["text"], self.train_data["author"]):
            text = re.split("[,.?，。？;\"\'\“\”\‘\’!！ ]",text) #利用正则表达式对句子分词
            for word in text:
                author_wordTable[author][self.wordTable.index(word)] += 1
        return author_wordTable
```

```python
naviebase = NaiveBaye("../data/train.csv")

print(naviebase.authorWordTable)
```
